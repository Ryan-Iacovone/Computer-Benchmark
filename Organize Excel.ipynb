```python
# Libraries 
import pandas as pd 
import os 
import xlsxwriter #engine for the pandas excel writer
import re #regular expression functions to parse file names
import sys
import datetime as dt #used to get todays date for naming frameview file
```

```python
#Where my list of excel files are located 
directory = r"C:\Users\Ryan\Documents\FrameView"

#name of singular output excel file
output_excel = r"C:\Users\Ryan\Coding Projects\Computer Benchmark\Data\frameview_files.xlsx" 

#Creating an empty list to fill with the names of the excel files I'm gonna merge into a single excel file  
excel_files = []

#looping through each file in the frameview directory and finding all the .csv files to merge into a single excel workbook
for file in os.listdir(directory):
    if file.endswith(".csv"):
        
        #splitting each csv file by the "_" then taking the 1st index of the created list to get the game name 
        new_file_name = re.split(string = file, pattern="_")[1]
        new_file_name = re.split(string = new_file_name, pattern= '\.')[0] #need the '\' to escape the period I guess 
        new_file_name = re.split(string = new_file_name, pattern= '-')[0]

        excel_files.append(new_file_name) 

        #renaming each of the existing files to their new simplied name I created above 
        os.rename(os.path.join(directory, file), os.path.join(directory, new_file_name + ".csv"))

#cleaning up the workspace
del new_file_name

#Creating a pandas writer object to write the excel file 
writer = pd.ExcelWriter(output_excel, engine = 'xlsxwriter')

#itterating over each newly created file to add it to the singualr excel sheet 
for file in excel_files:

    #Read the excel file
    file_path = os.path.join(directory, file + ".csv")
    df = pd.read_csv(file_path)

    #write each df to a new sheet in the output file
    sheet_name = os.path.splitext(file)[0] #using the file name as the name for the sheet
    df.to_excel(writer, sheet_name = sheet_name, index = False)

#save and close the writer 
writer.save()
writer.close()

#could also write code at the end of this to delete all of the orignal excel files in the frameview folder 

#Took 2 minutes 45 seconds to run 
#Now takes 4 minutes and 2 seconds to run with creating the excel file too
```

```python
#removing all (and only) the csv files from the frameview folder that we just merged together 

for file in os.listdir(directory):
    if file.endswith(".csv"):
        print(f"removing {file}..." )
        os.remove(os.path.join(directory, file))
    else:
        print("No more frameview created CSV files to remove :)")
```
